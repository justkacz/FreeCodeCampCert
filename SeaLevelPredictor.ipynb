```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
import scipy as sp
import numpy as np

df=pd.read_csv("https://raw.githubusercontent.com/justkacz/FreeCodeCampCert/main/epa-sea-level.csv")

y=df['CSIRO Adjusted Sea Level']
y2=df[df.Year>=2000]['CSIRO Adjusted Sea Level']
x=df.Year
x2=df[df.Year>=2000]['Year']

x_ext=np.arange(df.Year.min(), 2060, 10)
x_ext2=np.arange(2000, 2060, 1)
# x_ext2=np.append(x_ext2, np.array([2050]))

slope, intercept, r_value, p_value, stderr = linregress(x=x, y=y)
slope2, intercept2, r_value2, p_value2, stderr2 = linregress(x=x2, y=y2)

plt.plot(x, y, 'o')
plt.plot(x_ext, intercept + slope*x_ext, 'r', label = 'Whole period')
plt.plot(x_ext2, intercept2 + slope2*x_ext2, 'g', label = 'From the year 2000')
plt.xlabel("Year")
plt.ylabel("Sea Level (inches)")
plt.title("Rise in Sea Level")
plt.legend(loc="upper left")
```
